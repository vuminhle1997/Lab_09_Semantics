# Import polls and extract rapper artists from polls

```python
pollsPath = './data/polls.csv'
polls = pandas.read_csv(pollsPath)
artists = polls['artist'].unique().tolist()

polls
```

```python
actorFilmPath = './data/actorfilms.csv'
actorFilmlist = pandas.read_csv(actorFilmPath)
actors = actorFilmlist['Actor'].unique().tolist()

for i, a in enumerate(actors):
    print("{}: {}".format(i, a))
```

```python
# print(artists)
# print(actors)

# map artists and find the films they were in
overlappingData = []
for a in artists:
    films = actorFilmlist[actorFilmlist['Actor'] == a]['Film'].tolist()
    # print("---------")
    # print(a)
    # print(connection)

    # save overlapping data in a list
    if len(films) > 0:
        overlappingData.append({
            'name': a,
            'films': films
        })

# print(overlappingData)

for i, a in enumerate(overlappingData):
    print("{}: {}".format(i, a))
```

# Display singers and the contributions in films as bar chart

```python

counts = []
artistNames = []

for data in overlappingData:
    artistNames.append(data['name'])
    counts.append(len(data['films']))
    
plt.xticks(rotation=90)
plt.bar(artistNames, counts)
```

# Display artists and contributed films as table

```python
table = pandas.DataFrame(list(map(lambda x : (x['name'], len(x['films']), x['films']), overlappingData)))
table = table.rename(columns={0:'Artist',1:'Count', 2: 'Films'})
table
```

# Import movie database with the properties

```python
moviesData = pandas.read_csv('./data/movies_metadata.csv')
moviesData = moviesData[['title', 'id', 'imdb_id', 'original_title', 'vote_average', 'vote_count', 'revenue']]
moviesData.head(5)
```

# Import ratings of movies, aggregate and reduce by its score

```python
ratings = pandas.read_csv('./data/ratings.csv')

output = ratings.groupby('movieId').aggregate('rating')
output.head()
```

# Get movies in which the rap artists have played

```python
movies = pandas.DataFrame()

for data in overlappingData:
    for movie in data['films']:
        m = moviesData[moviesData['title'] == movie]
        if m.empty is False and m['title'] is not None:
            movies = pandas.concat([movies, m])

movies = movies.sort_values('title')
movies

            
```

# Visualize the average rating of the movies as a bar chart

```python
titles = []
score = []

for i, m in movies.iterrows():
    titles.append(m['title'])
    score.append(m['vote_average'])

plt.figure(figsize=(25, 5))
plt.xticks(rotation=90)
plt.bar(titles, score)
```

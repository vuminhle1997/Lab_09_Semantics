# Import polls and extract rapper artists from polls

```python
pollsPath = './data/polls.csv'
polls = pandas.read_csv(pollsPath)
artists = polls['artist'].unique().tolist()

for i, a in enumerate(artists):
    print("{}: {}".format(i, a))
```

```python
actorFilmPath = './data/actorfilms.csv'
actorFilmlist = pandas.read_csv(actorFilmPath)
actors = actorFilmlist['Actor'].unique().tolist()

for i, a in enumerate(actors):
    print("{}: {}".format(i, a))
```

```python
# print(artists)
# print(actors)

# map artists and find the films they were in
overlappingData = []
for a in artists:
    films = actorFilmlist[actorFilmlist['Actor'] == a]['Film'].tolist()
    # print("---------")
    # print(a)
    # print(connection)

    # save overlapping data in a list
    if len(films) > 0:
        overlappingData.append({
            'name': a,
            'films': films
        })

# print(overlappingData)

for i, a in enumerate(overlappingData):
    print("{}: {}".format(i, a))
```
